## A101 Factor 081

**Formula:**

`((rank(Log(product(rank((rank(correlation(vwap, sum(adv10, 49.6054), 8.47743))^4)), 14.9655))) < rank(correlation(rank(vwap), rank(volume), 5.07914))) * -1)`

```python
sql_alpha_081 = f"""--sql
SELECT
    date,
    instrument,
    IF((c_pct_rank(log(m_product(c_pct_rank((c_pct_rank(m_corr(vwap, m_sum(adv10, 49.6054), 8.47743))^4)), 14.9655))) 
         < c_pct_rank(m_corr(c_pct_rank(vwap), c_pct_rank(volume), 5.07914))), 1, 0) * -1  
    AS alpha_a101_f0081,
FROM data_base
"""
dai.query(get_a101_sql_base(sql_alpha_081), filters={'date': [sd, ed]}).df().dropna()
```

## A101 Factor 082

**Formula:**

`(min(rank(decay_linear(delta(open, 1.46063), 14.8717)), Ts_Rank(decay_linear(correlation(IndNeutralize(volume, IndClass.sector), ((open * 0.634196) + (open * (1 - 0.634196))), 17.4842), 6.92131), 13.4283)) * -1)`

## A101 Factor 083

**Formula:**

`((rank(delay(((high - low) / (sum(close, 5) / 5)), 2)) * rank(rank(volume))) / (((high - low) / (sum(close, 5) / 5)) / (vwap - close)))`

## A101 Factor 100

**Formula:**

`(0 - (1 * (((1.5 * scale(indneutralize(indneutralize(rank(((((close - low) - (high - close)) / (high - low)) * volume)), IndClass.subindustry), IndClass.subindustry))) - scale(indneutralize((correlation(close, rank(adv20), 5) - rank(ts_argmin(close, 30))), IndClass.subindustry))) * (volume / adv20))))`

```python
sql_alpha_100 = f"""--sql
SELECT
    date,
    instrument,
    (0 - (1 * (((1.5 * c_scale(c_indneutralize(
                                c_indneutralize(c_pct_rank(
                                    (((close - low) - (high - close)) / (high - low)) * volume) * 100,
                                class_lv3), class_lv3), 1))
                 - c_scale(c_indneutralize(
                        (m_corr(close, c_pct_rank(adv20), 5) - c_pct_rank(m_imin(close, 30))),
                        class_lv3), 1))
              * (volume / adv20))))
    AS alpha_a101_f0100,
FROM data_base
"""
dai.query(get_a101_sql_base(sql_alpha_100), filters={'date': [sd, ed]}).df().dropna()
```
