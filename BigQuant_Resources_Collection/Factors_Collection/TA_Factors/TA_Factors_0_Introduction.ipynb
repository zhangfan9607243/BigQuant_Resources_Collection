# Technical Factors

## 1. Technical Factors Introduction

Technical Factors can be applied to any financial securities at any data frequency.

To use these factors, users only need to specify the corresponding data table through the `data_base` parameter in each function.
The specified data table must contain the following columns: `open`, `close`, `high`, `low`, `volume`, `amount`. For stock-related factors, the turnover rate (`turn`) may be required. Some factors additionally require a benchmark data table specified by the `data_benchmark` parameter. Some securities or indicators provide only a single price (e.g., mid price, average price, aggregated value, etc.). These data can also be used to compute Technical Factors, although the set of supported factors is more limited. When performing the calculation, users should explicitly specify the single-price column name, or rename the single-price column to `close`.

## 2. Signal Types

Trend Signal:

- `TRBY`: Trend Upward   / Gold Cross / Buy
- `TRSL`: Trend Downward / Dead Cross / Sell

Momentum Signal:

- `MTBY`: Momentum Up   / Buy
- `MTSL`: Momentum Down / Sell

Reversal Signal:

- `RVBY`: Reverse Oversold   / Buy
- `RVSL`: Reverse Overbought / Sell

Breakout Signal:

- `BKBY`: Breaking Resistance / Buy
- `BKSL`: Breaking Support    / Sell

Volatility Signal:

- `VTUP`: Volatility Increase
- `VTDN`: Volatility Decrease

Volume Signal:

- `VMUP`: Volume expansion
- `VMDN`: Volume contraction
