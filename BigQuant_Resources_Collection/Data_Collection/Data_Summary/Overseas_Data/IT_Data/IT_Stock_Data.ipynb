# Italy Data - Stock Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. IT Stock Basic Information

```python
dai.query("""
SELECT *
FROM it_stock_basic_info
WHERE instrument = 'ISP.MI' 
""").df()
```

## 2. IT Stock Price-Volume & Trading Data

### 2.1 IT Stock Trading Record

```python
dai.query("""
SELECT *
FROM it_stock_instruments
WHERE date = '2025-01-31'
AND instrument = 'ISP.MI' 
""").df()
```

### 2.2 IT Stock Price-Volume Data

```python
dai.query("""
SELECT *
FROM it_stock_bar1d
WHERE date = '2025-01-31'
AND instrument = 'ISP.MI' 
""").df()
```
