# Japanese Data - Stock Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. JP Stock Basic Information

```python
dai.query("""
SELECT *
FROM jp_stock_basic_info
WHERE instrument = '6758.T' 
""").df()
```

## 2. JP Stock Price-Volume & Trading Data

### 2.1 JP Stock Trading Record

```python
dai.query("""
SELECT *
FROM jp_stock_instruments
WHERE date = '2025-01-31'
AND instrument = '6758.T'
""").df()
```

### 2.2 JP Stock Price-Volume Data

```python
dai.query("""
SELECT *
FROM jp_stock_bar1d
WHERE date = '2025-01-31'
AND instrument = '6758.T'
""").df()
```
