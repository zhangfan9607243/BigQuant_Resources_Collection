# Singapore Data - Stock Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. SG Stock Basic Information

```python
dai.query("""
SELECT *
FROM sg_stock_basic_info
WHERE instrument = 'J36.SI' 
""").df()
```

## 2. SG Stock Price-Volume & Trading Data

### 2.1 SG Stock Trading Record

```python
dai.query("""
SELECT *
FROM sgx_stock_instruments
WHERE date = '2023-12-29'
AND instrument = 'J36.SI' 
""").df()
```

### 2.2 SG Stock Price-Volume Data - Post-Adjusted

```python
dai.query("""
SELECT *
FROM sgx_stock_bar1d
WHERE date = '2023-12-29'
AND instrument = 'J36.SI' 
""").df()
```

### 2.3 SG Stock Price-Volume Data - Pre-Adjusted

```python
dai.query("""
SELECT *
FROM sg_stock_real_bar1d
WHERE date = '2024-07-02'
AND instrument = 'J36.SI' 
""").df()
```

## 3. SG Stock Status

```python
dai.query("""
SELECT *
FROM sg_stock_status
WHERE date = '2024-07-02'
AND instrument = 'J36.SI' 
""").df()
```
