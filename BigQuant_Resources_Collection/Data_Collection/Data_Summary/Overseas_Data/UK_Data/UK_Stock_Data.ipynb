# United Kingdom Data - Stock Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. UK Stock Basic Information

```python
dai.query("""
SELECT *
FROM uk_stock_basic_info
WHERE instrument = 'FERG.L' 
""").df()
```

## 2. UK Stock Price-Volume & Trading Data

### 2.1 UK Stock Trading Record

```python
dai.query("""
SELECT *
FROM uk_stock_instruments
WHERE date = '2025-01-31'
AND instrument = 'FERG.L' 
""").df()
```

### 2.2 UK Stock Price-Volume Data

```python
dai.query("""
SELECT *
FROM uk_stock_bar1d
WHERE date = '2025-01-31'
AND instrument = 'FERG.L' 
""").df()
```
