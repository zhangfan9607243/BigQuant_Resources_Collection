# Chinese Hong Kong Data - Stock Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. CNHK Stock Basic Information

```python
dai.query("""
SELECT *
FROM hk_stock_basic_info
WHERE instrument LIKE '%0700.HK%'
""").df()
```

## 2. CNHK Stock Price-Volume & Trading Data

### 2.1 CNHK Stock Trading Record

```python
dai.query("""
SELECT *
FROM cnhk_stock_instruments
WHERE date = '2025-06-30'
AND instrument LIKE '%0700.HK%'
""").df()
```

### 2.2 CNHK Stock Price-Volume Data - Post-Adjusted

```python
dai.query("""
SELECT *
FROM cnhk_stock_bar1d
WHERE date = '2025-06-30'
AND instrument LIKE '%0700.HK%'
""").df()
```

### 2.3 CNHK Stock Price-Volume Data - Pre-Adjusted

```python
dai.query("""
SELECT *
FROM hk_stock_real_bar1d
WHERE date = '2025-06-30'
AND instrument LIKE '%0700.HK%'
""").df()
```

## 2.4 CNHK Stock Price Comparison - Chinese Mainland vs Chinese Hong Kong

```python
dai.query("""
SELECT *
FROM hk_stock_ah_midpoint
WHERE date = '2025-06-30'
AND name LIKE '%平安%'
""").df()
```
