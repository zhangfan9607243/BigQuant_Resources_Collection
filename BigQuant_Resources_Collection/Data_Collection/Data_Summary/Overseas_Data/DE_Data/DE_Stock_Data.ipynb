# Germany Data - Stock Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. DE Stock Basic Information

```python
dai.query("""
SELECT *
FROM de_stock_basic_info
WHERE instrument = 'C6D0.F' 
""").df()
```

## 2. DE Stock Price-Volume & Trading Data

### 2.1 DE Stock Trading Record

```python
dai.query("""
SELECT *
FROM de_stock_instruments
WHERE date = '2025-01-31'
AND instrument = 'C6D0.F' 
""").df()
```

### 2.2 DE Stock Price-Volume Data

```python
dai.query("""
SELECT *
FROM de_stock_bar1d
WHERE date = '2025-01-31'
AND instrument = 'C6D0.F' 
""").df()
```
