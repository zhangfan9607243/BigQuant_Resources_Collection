# United States Data - Bond Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. US TBond Interest Rate

### 1.1 US TBond Short-Term Interest Rate

```python
dai.query("""
SELECT *
FROM us_tbond_short_rate
WHERE date = '2024-07-01'
""").df()
```

### 1.2 US TBond Long-Term Interest Rate

#### 1.2.1 Long-Term Interest Rate Data

```python
dai.query("""
SELECT *
FROM us_tbond_long_rate
WHERE date = '2024-07-01'
""").df()
```

#### 1.2.2 Long-Term Real Interest Rate Data

```python
dai.query("""
SELECT *
FROM us_tbond_real_long_rate_ave
WHERE date = '2024-07-01'
""").df()
```
