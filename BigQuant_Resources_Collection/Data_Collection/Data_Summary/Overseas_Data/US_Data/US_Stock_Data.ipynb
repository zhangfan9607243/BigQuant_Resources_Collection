# United States Data - Stock Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. US Stock Basic Information

```python
dai.query("""
SELECT *
FROM us_stock_basic_info
WHERE instrument = 'AAPL.O'
""").df()
```

## 2. US Stock Price-Volume & Trading Data

### 2.1 US Stock Trading Record

```python
dai.query("""
SELECT *
FROM us_stock_instruments
WHERE date = '2026-02-12'
AND instrument = 'AAPL.O'
""").df()
```

### 2.2 US Stock Price-Volume Data - Post-Adjusted

```python
dai.query("""
SELECT *
FROM us_stock_bar1d
WHERE date = '2026-02-12'
AND instrument = 'AAPL.O'
""").df()
```

### 2.3 US Stock Price-Volume Data - Pre-Adjusted

```python
dai.query("""
SELECT *
FROM us_stock_real_bar1d
WHERE date = '2026-02-12'
AND instrument = 'AAPL.O'
""").df()
```

## 3. US Stock Shares Data

```python
dai.query("""
SELECT *
FROM us_stock_shares
WHERE date = '2026-02-12'
AND instrument = 'AAPL.O'
""").df()
```
