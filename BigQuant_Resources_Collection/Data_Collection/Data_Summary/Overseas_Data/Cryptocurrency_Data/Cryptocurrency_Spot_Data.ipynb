# Cryptocurrency Data - Spot Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. Cryptocurrency Spot Price

### 1.1 Cryptocurrency Spot Price - Daily

```python
dai.query("""
SELECT *
FROM cc_binance_spot_bar1d
WHERE date = '2026-02-13'
AND instrument = 'BTCUSD'
""").df()
```

### 1.2 Cryptocurrency Spot Price - Minute

```python
dai.query("""
SELECT *
FROM cc_binance_spot_bar1m
WHERE date BETWEEN '2026-02-13 00:00:00.000000' AND '2026-02-13 23:59:59.999999'
AND instrument = 'BTCUSD'
""").df()
```

## 2. Cryptocurrency Spot Market Exchange Info & Trading Rules
