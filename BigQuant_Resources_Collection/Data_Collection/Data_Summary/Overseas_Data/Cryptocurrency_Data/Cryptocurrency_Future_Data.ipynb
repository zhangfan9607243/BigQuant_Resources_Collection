# Cryptocurrency Data - Future Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. Cryptocurrency USDT-M Futures

### 1.1 Cryptocurrency USDT-M Futures Metrics

```python
dai.query("""
SELECT *
FROM cc_binance_future_um_metrics_5m
WHERE date BETWEEN '2026-02-13 00:00:00.000000' AND '2026-02-13 23:59:59.999999'
AND instrument LIKE '%BTCUSD%'
""").df()
```

### 1.2 Cryptocurrency USDT-M Futures Market Exchange Info & Trading Rules

```python
dai.query("""
SELECT *
FROM cc_binance_future_um_exchange_info
WHERE instrument LIKE '%BTCUSD%'
""").df()
```
