# BigQuant Data Summary

This notebook summarizes the data resources currently supported by the BigQuant data platform.

## 1. Chinese Financial Market Data

### (1) Chinese Financial Instruments

The BigQuant data platform provides comprehensive coverage of financial instruments across Chinese financial markets, including:

Stock:

- Stock Securities:

    - Market Data:

        - Daily Frequency:

            - `cn_stock_instruments`: The stock codes and company names of equities traded daily in the China A-share market.
            - `cn_stock_bar1d`: Daily post-adjusted (back-adjusted) OHLCVA data, along with derived fields.
            - `cn_stock_real_bar1d`: Daily pre-adjusted (back-adjusted) OHLCVA data, along with derived fields.
            - `cn_stock_limit_price`: Daily upper and lower price limit information for each stock.
            - `cn_stock_wap`: Time-weighted average price (TWAP) and volume-weighted average price (VWAP) across different intraday time intervals.
            - `cn_stock_moneyflow_ns`: Minute-level capital flow data for cross-border stock trading conducted through the Shanghai-Hong Kong Stock Connect and Shenzhen-Hong Kong Stock Connect mechanisms, which includes both southbound flows (from Mainland China to Hong Kong) and northbound flows (from Hong Kong to Mainland China).
            - `cn_stock_static_data`: Pre-market daily stock status information for each trading day, e.g., trading suspension/resumption status, daily price limits, ST status, ex-rights/ex-dividend status, etc.

        - High Frequency:

            - `cn_stock_bar1m`: Daily-frequency OHLCVA data and derived fields.

    - Company & Security Basic Information

    - Company Financial Data
