# BigQuant Data Summary

## 1. Chinese (Mainland) Financial Market Data

### (1) Chinese (Mainland) Financial Instruments Data

The BigQuant data platform provides comprehensive coverage of financial instruments across Chinese (Mainland) financial markets, including:

- Stock:

    - Stock Securities Data

    - Stock Index Data
    
    - Stock Industry Data

- Bond

- Convertible Bond

- Exchange Traded Fund

- Future

- Option

- Other Data: Trading Calendar
