# Chinese (Mainland) Stock Data - Financial Data - Derived FS Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', None)
pd.set_option('display.max_column', None)
```

## 1. Financial Statements Derived Metrics

### 1.1 Financial Statements Derived Metrics - Latest Report

```python
dai.query("""
SELECT *
FROM cn_stock_financial_lf_shift
WHERE date >= '2024-12-31'
AND report_date = '2024-12-31'
AND instrument = '600519.SH'
AND shift = 0
""", full_db_scan=True).df().transpose()
```

### 1.2 Financial Statements Derived Metrics - Latest Annual Report

```python
dai.query("""
SELECT *
FROM cn_stock_financial_ly_shift
WHERE date >= '2024-12-31'
AND report_date = '2024-12-31'
AND instrument = '600519.SH'
AND shift = 0
LIMIT 1
""", full_db_scan=True).df().transpose()
```

### 1.3 Financial Statements Derived Metrics - Latest Quarter Report

```python
dai.query("""
SELECT *
FROM cn_stock_financial_mrq_shift
WHERE date >= '2024-12-31'
AND report_date = '2024-12-31'
AND instrument = '600519.SH'
AND shift = 0
""", full_db_scan=True).df().transpose()
```

### 1.4 Financial Statements Derived Metrics - Trailing Twelve Months

```python
dai.query("""
SELECT *
FROM cn_stock_financial_ttm_shift
WHERE date >= '2024-12-31'
AND report_date = '2024-12-31'
AND instrument = '600519.SH'
AND shift = 0
""", full_db_scan=True).df().transpose()
```
