# Chinese (Mainland) Stock Industry Data - Market Data - Price-Volume Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. Stock Industry Price-Volume Data

### 1.1 Post-Adjusted Daily Price-Volume Record

```python
# CS Industry Classification
dai.query("""
SELECT *
FROM cn_stock_industry_bar1d
WHERE date = '2026-02-13'
AND instrument = '101010'
""").df()
```

```python
# SW Industry Classification
dai.query("""
SELECT *
FROM cn_stock_industry_sw_bar1d
WHERE date = '2026-02-13'
AND instrument = '801010.SWI'
""").df()
```

### 1.2 Pre-Adjusted Daily Price-Volume Record

```python
# CS Industry Classification
dai.query("""
SELECT *
FROM cn_stock_industry_real_bar1d
WHERE date = '2026-02-13'
AND instrument = '101010'
""").df()
```
