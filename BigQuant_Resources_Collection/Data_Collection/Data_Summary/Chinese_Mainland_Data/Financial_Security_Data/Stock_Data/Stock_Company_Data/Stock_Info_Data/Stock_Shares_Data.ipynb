# Chinese (Mainland) Stock Data - Information Data - Shares Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 10)
pd.set_option('display.max_column', None)
```

## 1. Shares Data

### 1.1 Daily Shares Data

```python
dai.query("""
SELECT *
FROM cn_stock_shares
WHERE date = '2026-02-13'
AND instrument = '600519.SH'
""").df()
```

### 1.2 Historical Shares Data

#### 1.2.1 Shares Allotment

```python
dai.query("""
SELECT *
FROM cn_stock_allotment
WHERE instrument = '600601.SH'
ORDER BY date DESC, instrument
LIMIT 1
""").df()
```

#### 1.2.2 Shares Capital

```python
dai.query("""
SELECT *
FROM cn_stock_capital
WHERE instrument = '600519.SH'
ORDER BY publish_date DESC, instrument
LIMIT 1
""").df()
```

## 2. Dividend Data

```python
dai.query("""
SELECT *
FROM cn_stock_dividend
WHERE instrument = '600519.SH'
ORDER BY date DESC, instrument
LIMIT 1
""").df()
```

## 3. Valuation Data

```python
dai.query("""
SELECT *
FROM cn_stock_valuation
WHERE date = '2026-02-13'
AND instrument = '600519.SH'
""").df()
```

## 4. Shareholder Information

### 4.1 Shareholder Number & Changes

```python
dai.query("""
SELECT *
FROM cn_stock_shareholder
WHERE instrument = '600519.SH'
ORDER BY publish_date DESC, instrument
LIMIT 1
""").df()
```

```python
dai.query("""
SELECT *
FROM cn_stock_shareholder_change
WHERE instrument = '600519.SH'
ORDER BY end_date DESC, instrument
LIMIT 1
""").df()
```

### 4.2 Top Shareholders & Top Float Shareholders 

```python
dai.query("""
SELECT *
FROM cn_stock_shareholder_top10
WHERE instrument = '600519.SH'
ORDER BY publish_date DESC, instrument
LIMIT 1
""").df()
```

```python
dai.query("""
SELECT *
FROM cn_stock_top10_float_shareholder
WHERE instrument = '600519.SH'
ORDER BY publish_date DESC, instrument, rank
LIMIT 10
""").df()
```
