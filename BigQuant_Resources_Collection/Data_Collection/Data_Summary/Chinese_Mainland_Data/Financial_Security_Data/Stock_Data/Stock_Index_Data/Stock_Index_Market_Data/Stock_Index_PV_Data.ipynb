# Chinese (Mainland) Stock Index Data - Market Data - Price-Volume Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. Stock Index Data

### 1.1 Daily Price-Volume Record

```python
dai.query("""
SELECT *
FROM cn_stock_index_bar1d
WHERE date = '2026-02-13'
AND instrument = '000300.SH'
""").df()
```

### 1.2 Intraday Price-Volume Data

```python
dai.query("""
SELECT *
FROM cn_stock_index_bar1m
WHERE date = '2026-02-13 10:30:00'
AND instrument = '000300.SH'
""").df()
```
