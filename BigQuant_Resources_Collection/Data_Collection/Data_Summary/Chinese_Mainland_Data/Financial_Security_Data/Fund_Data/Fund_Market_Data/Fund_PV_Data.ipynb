# Chinese (Mainland) ETF Data - Market Data - Price-Volume Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. Daily Price-Volume & Trading Data

### 1.1 Daily Trading Record

```python
dai.query("""
SELECT *
FROM cn_fund_instruments
WHERE date = '2026-02-13'
AND instrument = '588990.SH'
""").df()
```

### 1.2 Post-Adjusted Price-Volume Data

```python
dai.query("""
SELECT *
FROM cn_fund_bar1d
WHERE date = '2026-02-13'
AND instrument = '588990.SH'
""").df()
```

### 1.3 Pre-Adjusted Price-Volume Data

```python
dai.query("""
SELECT *
FROM cn_fund_real_bar1d
WHERE date = '2026-02-13'
AND instrument = '588990.SH'
""").df()
```

## 2. Intraday Price-Volume & Trading Data

```python
dai.query("""
SELECT *
FROM cn_fund_bar1m
WHERE date BETWEEN '2026-02-13 00:00:00.000' AND '2026-02-13 23:59:59.999'
AND instrument = '588990.SH'
""").df()
```
