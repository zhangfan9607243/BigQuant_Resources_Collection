# Chinese (Mainland) Convertible Bond Data - Information Data - Bond Covenants

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. Convertible Bond Basic Information

```python
dai.query("""
SELECT *
FROM cn_cbond_basic_info
WHERE instrument = '127040.SZ'
""").df()
```

## 2. Convertible Bond Coupon Rate

```python
dai.query("""
SELECT *
FROM cn_cbond_coupon_rate_te
WHERE instrument = '127040.SZ'
""").df()
```

## 3. Convertible Bond Issue & Repurchase & Conversion

### 3.1 Convertible Bond Issue

```python
dai.query("""
SELECT *
FROM cn_cbond_issue_te
WHERE instrument = '127040.SZ'
""").df()
```

### 3.2 Convertible Bond Repurchase

```python
dai.query("""
SELECT *
FROM cn_cbond_repurchase_bar1d_te
WHERE date > '2024-06-30'
AND instrument = '131800.SZ'
""").df()
```

### 3.3 Convertible Bond Conversion

```python
dai.query("""
SELECT *
FROM cn_cbond_conversion_result_te
WHERE publish_date > '2015-01-01'
AND instrument = '113001.SH' 
""").df()
```
