# Chinese (Mainland) Future Data - Information Data - Basic Information

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. Basic Information Data

### 1.1 Product Information

```python
dai.query("""
SELECT *
FROM cn_future_product_info
WHERE code = 'rb'
""").df()
```

### 1.2 Future Information

```python
dai.query("""
SELECT *
FROM cn_future_basic_info
WHERE instrument = 'rb2612.SHF'
""").df()
```

## 2. Future Domainant Contract

```python
dai.query("""
SELECT *
FROM cn_future_dominant
WHERE date = '2026-02-13'
AND product_code = 'rb'
""").df()
```

## 3. Warehouse Data

```python
dai.query("""
SELECT *
FROM cn_future_warehouse_receipt
WHERE date = '2025-07-11'
AND product_code = 'rb'
""").df()
```
