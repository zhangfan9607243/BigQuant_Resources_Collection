# Chinese (Mainland) Future Data - Market Data - Price-Volume Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. Daily Price-Volume & Trading Data

## 1.1 Daily Trading Record

```python
dai.query("""
SELECT *
FROM cn_future_instruments
WHERE date = '2026-02-13'
AND instrument = 'rb2612.SHF'
""").df()
```

### 1.2 Pre-Adjusted Price-Volume Data

```python
dai.query("""
SELECT *
FROM cn_future_bar1d
WHERE date = '2026-02-13'
AND instrument = 'rb2612.SHF'
""").df()
```

### 1.3 Post-Adjusted Price-Volume Data

```python
dai.query("""
SELECT *
FROM cn_future_bar1d_adjust
WHERE date = '2026-02-13'
AND product_code = 'rb'
""").df()
```

```python
dai.query("""
SELECT *
FROM cn_future_adjust_factor
WHERE date = '2025-09-01'
AND product_code = 'rb'
""").df()
```

## 2. Intraday Price-Volume & Trading Data

### 2.1 1-min Price-Volume Data

```python
dai.query("""
SELECT *
FROM cn_future_bar1m
WHERE date BETWEEN '2026-02-12 00:00:00.000' AND '2026-02-13 23:59:59.999'
AND trading_day = '20260213'
AND instrument = 'rb2612.SHF'
""").df()
```
