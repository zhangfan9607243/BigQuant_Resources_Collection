# Chinese (Mainland) Option Data - Market Data - Price-Volume Data

```python
import dai
import pandas as pd
pd.set_option('display.max_row', 6)
pd.set_option('display.max_column', None)
```

## 1. Daily Price-Volume & Trading Data

```python
dai.query("""
SELECT *
FROM cn_option_bar1d
WHERE date = '2026-02-13'
AND instrument = '10009633.SHO'
""").df()
```

## 2. Intraday Price-Volume & Trading Data

### 2.1 Intraday Price Data

```python
dai.query("""
SELECT *
FROM cn_option_bar1m
WHERE date BETWEEN '2026-02-13 00:00:00.000' AND '2026-02-13 23:59:59.999'
AND instrument = '10009633.SHO'
""").df()
```

### 2.2 Intraday VIX Data

```python
dai.query("""
SELECT *
FROM cn_option_vix_bar1m
WHERE date BETWEEN '2026-02-13 00:00:00.000' AND '2026-02-13 23:59:59.999'
AND instrument = '159901.VIX'
""").df()
```
