# 1. Car Market

### 1.1 Car Market Overview

```python
dai.query("""
SELECT *
FROM car_market_total_cpca
WHERE yearly_comparison_value = '2023年12月'
""").df()
```

### 1.2 Car Market Segmentation

#### 1.2.1 Car Market Segmentation by Powertrain Type

```python
dai.query("""
SELECT *
FROM car_market_fuel_cpca
WHERE yearly_comparison_value = '2023年12月'
""").df()
```

#### 1.2.2 Car Market Segmentation by Vehicle Category

```python
dai.query("""
SELECT *
FROM car_market_cate_cpca
WHERE yearly_comparison_value = '2023年12月'
""").df()
```

#### 1.2.3 Car Market Segmentation by Vehicle Segment

```python
dai.query("""
SELECT *
FROM car_market_segment_cpca
WHERE date = '2023年12月'
""").df()
```

#### 1.2.3 Car Market Segmentation by Country

```python
dai.query("""
SELECT *
FROM car_market_country_cpca
WHERE date = '2023年12月'
""").df()
```

### 1.3 Market Competition Data

```python
dai.query("""
SELECT *
FROM car_market_man_rank_cpca
WHERE date_comparison_value = '2023年12月'
""").df()
```
