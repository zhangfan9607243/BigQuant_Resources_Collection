```python
import dai
import pandas as pd
from datetime import datetime, timedelta
pd.set_option('display.max_rows', None)
```

```python
sql = """
WITH 
data_base AS (
    SELECT DISTINCT date
    FROM cn_stock_bar1m_c
    WHERE '2024-11-19 00:00:00' <= date AND date < '2024-11-20 00:00:00'
)
SELECT 
    date,
    ROW_NUMBER(date) AS date_index_minute,
FROM data_base
ORDER BY date
"""

df = dai.query(sql).df()
```

```python
df["dt_session_trade"]   = df["date"].apply(lambda x: f"{x.strftime('%Y%m%d')}{'AM' if x.hour < 12 else 'PM'}")
df
```

```python
def get_dt_hour_trade(dt):
    if dt.time() >= pd.Timestamp("09:00").time() and dt.time() <= pd.Timestamp("10:30").time():
        return f"{dt.strftime('%Y%m%d')}1"
    elif dt.time() > pd.Timestamp("10:30").time() and dt.time() <= pd.Timestamp("11:30").time():
        return f"{dt.strftime('%Y%m%d')}2"
    elif dt.time() > pd.Timestamp("13:00").time() and dt.time() <= pd.Timestamp("14:00").time():
        return f"{dt.strftime('%Y%m%d')}3"
    elif dt.time() > pd.Timestamp("14:00").time() and dt.time() <= pd.Timestamp("15:00").time():
        return f"{dt.strftime('%Y%m%d')}4"
    else:
        return None

df["dt_hour_trade"] = df["date"].apply(get_dt_hour_trade)
df
```
