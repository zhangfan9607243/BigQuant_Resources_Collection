```python
import dai
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
sql = """
WITH 
data_base AS (
    SELECT DISTINCT date
    FROM cn_future_bar1m_c
    WHERE '2024-11-19 00:00:00' <= date AND date < '2024-11-20 00:00:00'
)
SELECT 
    date,
    ROW_NUMBER(date) AS date_index_minute,
FROM data_base
ORDER BY date
"""

df = dai.query(sql).df()
df
```
