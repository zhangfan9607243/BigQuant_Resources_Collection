```python
import dai
import pandas as pd
```

```python
sql = """

WITH 
data_base AS (
    SELECT date
    FROM all_trading_days
    WHERE market_code = 'CN'
    QUALIFY date >= '1991-01-01'
)

SELECT 

    date,
    ROW_NUMBER(date) AS date_index,

    CONCAT(LPAD(CAST(year(date) AS CHAR), 4, '0')) AS dt_year,
    CONCAT(LPAD(CAST(year(date) AS CHAR), 4, '0'), LPAD(CAST(quarter(date) AS CHAR), 2, '0')) AS dt_quarter,
    CONCAT(LPAD(CAST(year(date) AS CHAR), 4, '0'), LPAD(CAST(month(date) AS CHAR), 2, '0')) AS dt_month,
    CONCAT(LPAD(CAST(year(date) AS CHAR), 4, '0'), LPAD(CAST(week(date) AS CHAR), 2, '0'))  AS dt_week,
    CONCAT(LPAD(CAST(year(date) AS CHAR), 4, '0'), LPAD(CAST(month(date) AS CHAR), 2, '0'), LPAD(CAST(day(date) AS CHAR), 2, '0')) AS dt_day,
    -- CONCAT(LPAD(CAST(year(date) AS CHAR), 4, '0'), LPAD(CAST(month(date) AS CHAR), 2, '0'), LPAD(CAST(day(date) AS CHAR), 2, '0'), LPAD(CAST(hour(date) AS CHAR), 2, '0')) AS dt_hour,
    -- CONCAT(LPAD(CAST(year(date) AS CHAR), 4, '0'), LPAD(CAST(month(date) AS CHAR), 2, '0'), LPAD(CAST(day(date) AS CHAR), 2, '0'), LPAD(CAST(hour(date) AS CHAR), 2, '0'), LPAD(CAST(minute(date) AS CHAR), 2, '0')) AS dt_minute,
    -- CONCAT(LPAD(CAST(year(date) AS CHAR), 4, '0'), LPAD(CAST(month(date) AS CHAR), 2, '0'), LPAD(CAST(day(date) AS CHAR), 2, '0'), LPAD(CAST(hour(date) AS CHAR), 2, '0'), LPAD(CAST(minute(date) AS CHAR), 2, '0'), LPAD(CAST(second(date) AS CHAR), 2, '0')) AS dt_second,
    -- CONCAT(LPAD(CAST(year(date) AS CHAR), 4, '0'), LPAD(CAST(month(date) AS CHAR), 2, '0'), LPAD(CAST(day(date) AS CHAR), 2, '0'), LPAD(CAST(hour(date) AS CHAR), 2, '0'), LPAD(CAST(minute(date) AS CHAR), 2, '0'), LPAD(CAST(second(date) AS CHAR), 2, '0'), LPAD(CAST(millisecond(date) AS CHAR), 3, '0')) AS dt_millisecond,
    -- CONCAT(LPAD(CAST(year(date) AS CHAR), 4, '0'), LPAD(CAST(month(date) AS CHAR), 2, '0'), LPAD(CAST(day(date) AS CHAR), 2, '0'), LPAD(CAST(hour(date) AS CHAR), 2, '0'), LPAD(CAST(minute(date) AS CHAR), 2, '0'), LPAD(CAST(second(date) AS CHAR), 2, '0'), LPAD(CAST(millisecond(date) AS CHAR), 3, '0'), LPAD(CAST(microsecond(date) AS CHAR), 3, '0')) AS dt_microsecond,
    
    IF(dt_year != LAG(dt_year, 1)  OVER (ORDER BY date), 1, 0) AS is_year_start,
    IF(dt_year != LAG(dt_year, -1) OVER (ORDER BY date), 1, 0) AS is_year_end,
    IF(dt_quarter != LAG(dt_quarter, 1)  OVER (ORDER BY date), 1, 0) AS is_quarter_start,
    IF(dt_quarter != LAG(dt_quarter, -1) OVER (ORDER BY date), 1, 0) AS is_quarter_end,
    IF(dt_month != LAG(dt_month, 1)  OVER (ORDER BY date), 1, 0) AS is_month_start,
    IF(dt_month != LAG(dt_month, -1) OVER (ORDER BY date), 1, 0) AS is_month_end,
    IF(dt_week != LAG(dt_week, 1)  OVER (ORDER BY date), 1, 0) AS is_week_start,
    IF(dt_week != LAG(dt_week, -1) OVER (ORDER BY date), 1, 0) AS is_week_end,

    
    ROW_NUMBER() OVER (PARTITION BY dt_year    ORDER BY date) AS day_of_year_trading,
    ROW_NUMBER() OVER (PARTITION BY dt_quarter ORDER BY date) AS day_of_quarter_trading,
    ROW_NUMBER() OVER (PARTITION BY dt_month   ORDER BY date) AS day_of_month_trading,
    ROW_NUMBER() OVER (PARTITION BY dt_week    ORDER BY date) AS day_of_week_trading,
    
    dayofyear(date)  AS day_of_year_nature,
    dayofmonth(date) AS day_of_month_nature,
    dayofweek(date)  AS day_of_week_nature,

FROM data_base

ORDER BY date

"""

df = dai.query(sql).df()
df
```

```python
df['day_of_quarter_nature'] = df['date'].apply(lambda x: (x - x.to_period('Q').start_time).days + 1)
df
```

```python
ds = dai.DataSource("mldt_cn_trading_calendar")

def update_df(df_new):
    df_new = df
    return df_new

ds.apply_bdb(update_df, as_type=pd.DataFrame)
```

```python
dai.DataSource.write_bdb(
    data=df,
    id="mldt_cn_trading_calendar",
    unique_together=["date"],
    indexes=["date"],
)
```
